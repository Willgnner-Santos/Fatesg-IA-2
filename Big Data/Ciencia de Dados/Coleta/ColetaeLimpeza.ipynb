```
import pandas as pd

# Passo 1: Carregar os dados
# Dataset: https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv
df_titanic = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")

# ConfiguraÃ§Ãµes de exibiÃ§Ã£o do Pandas para mostrar todas as colunas
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

# Passo 2: Visualizar as 5 primeiras linhas do dataset
print("\nğŸ” Primeiras linhas:")
print(df_titanic.head())

# Passo 3: Verificar o nome das colunas e tipos de dados
print("\nğŸ§¾ InformaÃ§Ãµes do DataFrame:")
print(df_titanic.info())

# EstatÃ­sticas descritivas para colunas numÃ©ricas
print("\nEstatÃ­sticas Descritivas para colunas numÃ©ricas:")
print(df_titanic.describe())

# Contagem de valores para uma coluna categÃ³rica (por exemplo, 'Survived')
print("\nContagem de sobreviventes (0 = NÃ£o, 1 = Sim):")
print(df_titanic['Survived'].value_counts())

# Passo 4: Filtrar apenas os passageiros que sobreviveram
sobreviventes = df_titanic[df_titanic['Survived'] == 1]
print("\nğŸ›Ÿ Passageiros que sobreviveram:")
print(sobreviventes.head())

# Passo 5: Criar um novo DataFrame com apenas colunas relevantes
resumo = df_titanic[['Name', 'Sex', 'Age', 'Survived']]
print("\nğŸ‘¤ Resumo com nome, sexo, idade e sobrevivÃªncia:")
print(resumo.head())

# Passo 6: Calcular a mÃ©dia de idade dos passageiros
media_idade = df_titanic['Age'].mean()
print(f"\nğŸ“Š MÃ©dia de idade dos passageiros: {media_idade:.2f} anos")

# Passo 7: Filtrar passageiros do sexo feminino com menos de 18 anos
# Nota: Esta filtragem usa a coluna 'Age' que tem valores NaN no dataset original.
# O cÃ¡lculo da mÃ©dia de idade acima ignora os NaNs, mas o filtro os trata como False.
meninas = df_titanic[(df_titanic['Sex'] == 'female') & (df_titanic['Age'] < 18)]
print("\nğŸ‘§ Passageiras com menos de 18 anos:")
print(meninas[['Name', 'Age']])

tickets = df_titanic['Ticket']
print(f"\nTickets (Apenas as primeiras linhas para visualizaÃ§Ã£o):\n{tickets.head()}")

# =========================================================================
# === SOLUÃ‡Ã•ES PARA OS DESAFIOS EXTRA ===
# =========================================================================

print("\n" + "="*50)
print("=== RESULTADOS DOS DESAFIOS EXTRAS ===")
print("="*50)

# DESAFIO 1: Filtrar apenas os passageiros da 1Âª classe que nÃ£o sobreviveram
#            Calcular a porcentagem desses em relaÃ§Ã£o ao total da 1Âª classe

# 1. Filtrar passageiros da 1Âª classe
passageiros_p1 = df_titanic[df_titanic['Pclass'] == 1]
total_p1 = len(passageiros_p1)

# 2. Filtrar nÃ£o sobreviventes da 1Âª classe
nao_sobrev_p1 = passageiros_p1[passageiros_p1['Survived'] == 0]
num_nao_sobrev_p1 = len(nao_sobrev_p1)

# 3. Calcular a porcentagem
if total_p1 > 0:
    porcentagem_nao_sobrev_p1 = (num_nao_sobrev_p1 / total_p1) * 100
    print(f"\n--- 1Âª Classe: NÃ£o Sobreviventes ---")
    print(f"Total de passageiros na 1Âª classe: {total_p1}")
    print(f"NÃ£o sobreviventes na 1Âª classe: {num_nao_sobrev_p1}")
    print(f"Porcentagem de nÃ£o sobreviventes na 1Âª classe: {porcentagem_nao_sobrev_p1:.2f}%")
else:
    print("\nNÃ£o hÃ¡ passageiros na 1Âª classe para realizar o cÃ¡lculo.")


# DESAFIO 2: Quantos passageiros eram de cada sexo e de qual porto de embarque?

print("\n--- Contagem por Sexo e Porto de Embarque (Embarked) ---")
# Usando pd.crosstab() para uma visualizaÃ§Ã£o de tabela de contingÃªncia
contagem_sexo_embarque = pd.crosstab(df_titanic['Sex'], df_titanic['Embarked'])
print(contagem_sexo_embarque)


# DESAFIO 3: Qual Ã© a idade mÃ©dia dos sobreviventes e dos nÃ£o sobreviventes?

print("\n--- Idade MÃ©dia por Status de SobrevivÃªncia ---")
# Usamos groupby() na coluna 'Survived' e aplicamos a funÃ§Ã£o mean() na coluna 'Age'
media_idade_sobrevivencia = df_titanic.groupby('Survived')['Age'].mean()

print(f"Idade mÃ©dia dos NÃ£o Sobreviventes (0): {media_idade_sobrevivencia.get(0, 'N/A'):.2f} anos")
print(f"Idade mÃ©dia dos Sobreviventes (1): {media_idade_sobrevivencia.get(1, 'N/A'):.2f} anos")
```

```
import pandas as pd
import numpy as np

# Definindo as opÃ§Ãµes de exibiÃ§Ã£o para melhor visualizaÃ§Ã£o
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

# ----------------------------------------------------
# PASSO 1: Carregar os Dados
# ----------------------------------------------------
print("--- ğŸš¢ Carregamento e PreparaÃ§Ã£o dos Dados ---")
df_titanic = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df_limpo = df_titanic.copy() # Cria uma cÃ³pia para o prÃ©-processamento

# ----------------------------------------------------
# PASSO 2: Limpeza e PrÃ©-processamento
# (Baseado no seu cÃ³digo)
# ----------------------------------------------------
num_duplicatas = df_limpo.duplicated().sum()
if num_duplicatas == 0:
    print("Nenhuma linha duplicada encontrada.")

# 3. Tratamento de Valores Ausentes
# Tratar a coluna 'Age' (idade) com a mediana
mediana_idade = df_limpo['Age'].median()
df_limpo.loc[:, 'Age'] = df_limpo['Age'].fillna(mediana_idade)
print(f"Valores ausentes na coluna 'Age' preenchidos com a mediana: {mediana_idade:.2f}")

# 4. Tratar a coluna 'Embarked' (porto de embarque) com o valor mais frequente
moda_embarque = df_limpo['Embarked'].mode()[0]
df_limpo.loc[:, 'Embarked'] = df_limpo['Embarked'].fillna(moda_embarque)
print(f"Valores ausentes na coluna 'Embarked' preenchidos com a moda: {moda_embarque}")

# 5. Para a coluna 'Cabin' (cabine), que tem muitos valores ausentes, vamos apenas indicar sua ausÃªncia
df_limpo.loc[:, 'Cabin'] = df_limpo['Cabin'].fillna('Missing')
print("Valores ausentes na coluna 'Cabin' preenchidos com 'Missing'.")

print("\nVerificaÃ§Ã£o final de valores ausentes:")
print(df_limpo.isnull().sum())

# ----------------------------------------------------
# DESAFIOS EXTRAS E SEGUNDA PARTE (AnÃ¡lise)
# ----------------------------------------------------

print("\n\n--- ğŸ“Š Resultados da AnÃ¡lise de Dados ---")

## A. Desafios de SobrevivÃªncia/Classe

# 1. Filtrar apenas os passageiros da 1Âª classe que nÃ£o sobreviveram
nao_sobrev_p1 = df_limpo[(df_limpo['Pclass'] == 1) & (df_limpo['Survived'] == 0)]
num_nao_sobrev_p1 = len(nao_sobrev_p1)

# Calcular a porcentagem desses em relaÃ§Ã£o ao total da 1Âª classe
total_p1 = len(df_limpo[df_limpo['Pclass'] == 1])
porcentagem_nao_sobrev_p1 = (num_nao_sobrev_p1 / total_p1) * 100

print("\n1. AnÃ¡lise da 1Âª Classe NÃ£o Sobrevivente:")
print(f"Total de passageiros na 1Âª classe: **{total_p1}**")
print(f"NÃ£o sobreviventes na 1Âª classe: **{num_nao_sobrev_p1}**")
print(f"Porcentagem de nÃ£o sobreviventes na 1Âª classe: **{porcentagem_nao_sobrev_p1:.2f}%**")

# 2. Quantos passageiros eram de cada sexo e de qual porto de embarque?
print("\n2. Contagem por Sexo e Porto de Embarque (Embarked):")
contagem_sexo_embarque = pd.crosstab(df_limpo['Sex'], df_limpo['Embarked'])
print(contagem_sexo_embarque)

# 3. Qual Ã© a idade mÃ©dia dos sobreviventes e dos nÃ£o sobreviventes?
media_idade_sobrevivencia = df_limpo.groupby('Survived')['Age'].mean()

print("\n3. Idade MÃ©dia por Status de SobrevivÃªncia:")
print(f"Idade mÃ©dia dos NÃ£o Sobreviventes (0): **{media_idade_sobrevivencia[0]:.2f} anos**")
print(f"Idade mÃ©dia dos Sobreviventes (1): **{media_idade_sobrevivencia[1]:.2f} anos**")


## B. Desafios de Limpeza e FamilySize

# 4. AnÃ¡lise 'Ticket': Remover linhas com valores ausentes
df_dropna = df_titanic.dropna()
num_linhas_original = df_titanic.shape[0]
num_linhas_restantes = df_dropna.shape[0]
linhas_perdidas = num_linhas_original - num_linhas_restantes

print("\n4. Impacto da RemoÃ§Ã£o Simples de Valores Ausentes (`dropna`):")
print(f"Tamanho do dataset original: **{num_linhas_original} linhas**")
print(f"Tamanho do dataset apÃ³s `dropna()`: **{num_linhas_restantes} linhas**")
print(f"Linhas perdidas: **{linhas_perdidas} ({linhas_perdidas/num_linhas_original*100:.2f}%)**")
print("ObservaÃ§Ã£o: A perda de dados (cerca de 79%) Ã© muito alta, confirmando que a imputaÃ§Ã£o (feita em 'df_limpo') foi a melhor estratÃ©gia.")

# 5. Criar 'FamilySize' e calcular a mÃ©dia de sobrevivÃªncia
df_limpo.loc[:, 'FamilySize'] = df_limpo['SibSp'] + df_limpo['Parch'] + 1
media_sobrevivencia_por_familia = df_limpo.groupby('FamilySize')['Survived'].mean().sort_values(ascending=False)

print("\n5. MÃ©dia de SobrevivÃªncia por Tamanho da FamÃ­lia (`FamilySize`):")
print(media_sobrevivencia_por_familia)

print("\n**O que se observa sobre o Tamanho da FamÃ­lia?**")
print("* Melhores chances: FamÃ­lias de tamanho **2, 3 e 4** tiveram a maior taxa de sobrevivÃªncia (acima de 55%).")
print("* Piores chances: Passageiros **sozinhos** (FamilySize=1) e em famÃ­lias **muito grandes** (8 ou 11 membros) tiveram as piores taxas (30% ou 0%, respectivamente).")


# ----------------------------------------------------
# ANÃLISE DE FRAUDE (Seguradora)
# ----------------------------------------------------
print("\n\n--- ğŸ’° AnÃ¡lise de Fraude para Seguradoras ---")

# Exemplo de AnÃ¡lise de Fraude: InconsistÃªncia de Classe vs. Tarifa

print("O tipo de informaÃ§Ã£o mais importante para um gestor de seguros seria a detecÃ§Ã£o de **inconsistÃªncias que pudessem justificar a superavaliaÃ§Ã£o da indenizaÃ§Ã£o**.")
print("\nExemplo de Filtro de Fraude: Passagens de 3Âª Classe com Tarifa Muito Alta (Suspeita de SuperavaliaÃ§Ã£o)")

# Calcular o 3Âº quartil da tarifa da 2Âª Classe (um benchmark para tarifa 'mÃ©dia-alta')
q3_fare_p2 = df_limpo[df_limpo['Pclass'] == 2]['Fare'].quantile(0.75)
print(f"Tarifa do 3Âº quartil da 2Âª Classe (Q3): **${q3_fare_p2:.2f}**")

# Filtrar passageiros da 3Âª Classe que pagaram uma tarifa acima do 3Âº quartil da 2Âª Classe
suspeitos_fare_p3 = df_limpo[(df_limpo['Pclass'] == 3) & (df_limpo['Fare'] > q3_fare_p2)]

print("\nSuspeitos (3Âª Classe com Tarifa AtÃ­pica Alta):")
if not suspeitos_fare_p3.empty:
    print(suspeitos_fare_p3[['Name', 'Pclass', 'Fare', 'Survived']].head(10))
    print(f"\n**Alerta:** Encontrados {len(suspeitos_fare_p3)} passageiros da 3Âª Classe pagando tarifas que indicam discrepÃ¢ncia. Isso pode ser um alerta para fraude na declaraÃ§Ã£o de valor do bilhete para fins de seguro.")
else:
    print("Nenhum passageiro da 3Âª Classe encontrado pagando tarifa acima do Q3 da 2Âª Classe.")
```

Limpeza e PadrÃµes de SobrevivÃªncia

Limpeza: Os dados ausentes foram tratados usando a mediana (28.00) para Age e a moda (S) para Embarked. A remoÃ§Ã£o simples resultaria na perda de $79.46\%$ dos dados, o que foi evitado.

Idade e Classe: Os sobreviventes eram, em mÃ©dia, mais jovens ($\approx 28.3$ anos), e a Primeira Classe teve a maior taxa de resgate ($\approx 63\%$).

Tamanho da FamÃ­lia: A melhor chance de sobrevivÃªncia foi para famÃ­lias de tamanho 2, 3 ou 4 (acima de $55\%$ de taxa). Passageiros sozinhos ou em grupos muito grandes tiveram as piores chances.

Alertas de Fraude para Seguradoras
Para detectar fraudes, o gestor de seguros deve focar em inconsistÃªncias nos registros:

Rastreamento por Bilhete: Investigar reivindicaÃ§Ãµes de morte para grupos (bilhetes) que, estatisticamente, deveriam ter tido sobreviventes (principalmente famÃ­lias de 2 a 4).

Fraude de Valor: Filtrar passageiros da 3Âª Classe que pagaram tarifas atÃ­picas (muito altas). Isso sugere que o valor do bilhete foi inflacionado para obter uma indenizaÃ§Ã£o de seguro maior.

ContradiÃ§Ã£o no Perfil: Verificar rigorosamente as reivindicaÃ§Ãµes de morte para crianÃ§as ou Homens da 3Âª Classe, pois estes perfis contradizem os padrÃµes de resgate, indicando possÃ­vel fraude de identidade.
