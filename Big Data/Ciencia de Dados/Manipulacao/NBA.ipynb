```
#Pontos por jogo
df['PTS'].hist(bins=40)

plt.title('Distribuição de Pontos por Jogo')
plt.xlabel('Pontos')
plt.ylabel('Frequência')

plt.show()
sns.scatterplot(x='FGA', y='PTS', data=df, alpha=0.3)

#Arremessos e Pontos convertidos
plt.title('Correlação entre Arremessos Tentados (FGA) e Pontos (PTS)')
plt.xlabel('Arremessos Tentados (FGA)')
plt.ylabel('Pontos (PTS)')
plt.show()

#Pontos por temperada em média
df.groupby('SEASON')['PTS'].mean().plot(kind='line')

plt.title('Média de Pontos por Temporada')
plt.xlabel('Temporada')
plt.ylabel('Pontos Médios por Jogo')
plt.show()

```

```
#Pontos
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

# Estatísticas
media_pts = df['PTS'].mean()
mediana_pts = df['PTS'].median()
moda_pts = df['PTS'].mode()[0]

# Histograma
df['PTS'].hist(bins=40, edgecolor='black', alpha=0.7)

# Linhas verticals
plt.axvline(media_pts, color='red', linestyle='dashed', linewidth=2, label=f'Média: {media_pts:.2f}')
plt.axvline(mediana_pts, color='green', linestyle='solid', linewidth=2, label=f'Mediana: {mediana_pts:.2f}')
plt.axvline(moda_pts, color='purple', linestyle='dotted', linewidth=2, label=f'Moda: {moda_pts:.2f}')

# Títulos
plt.title('Distribuição dos Pontos Marcados por Jogo (PTS)', fontsize=16)
plt.xlabel('Pontos por jogo', fontsize=12)
plt.ylabel('Frequência', fontsize=12)

plt.legend()
plt.grid(axis='y', alpha=0.5)

plt.show()
```

```
# Média de pontos por time
media_pts_por_time = df.groupby('TEAM_NAME')['PTS'].mean().sort_values(ascending=False)

plt.figure(figsize=(14, 6))
media_pts_por_time.plot(kind='bar')

plt.title('Média de Pontos por Jogo (PTS) - Por Time')
plt.ylabel('Média de Pontos')
plt.xlabel('Time')

plt.xticks(rotation=90)
plt.grid(axis='y', alpha=0.4)

plt.show()
```

```
#Taxa de vitória de cada time

taxa_por_time = df.groupby('TEAM_NAME')['RESULT'].mean().sort_values()

plt.figure(figsize=(10, 12))

sns.barplot(
    data=taxa_por_time.reset_index(),
    x='RESULT',
    y='TEAM_NAME',
    hue='TEAM_NAME',
    palette='viridis',
    legend=False
)

plt.title('Taxa de Vitória por Time (Todas as Partidas)', fontsize=16)
plt.xlabel('Taxa de vitória')
plt.ylabel('Time')
plt.show()
```
