```python
import pandas as pd

caminho_arquivo = '/content/titanic.csv'

try:
    df = pd.read_csv(caminho_arquivo)

    print(f"✅ Arquivo {caminho_arquivo} lido com sucesso!")
    print(f"DataFrame criado com {df.shape[0]} linhas e {df.shape[1]} colunas.")

except FileNotFoundError:
    print("⚠️ Erro: Certifique-se de ter carregado o arquivo 'titanic.csv' antes de executar este bloco.")
```

```python
mapa_colunas = {
    'PassengerId': 'id_passageiro',
    'Survived': 'sobreviveu',
    'Pclass': 'classe_p',
    'Name': 'nome',
    'Sex': 'sexo',
    'Age': 'idade',
    'SibSp': 'irmao_conjuge',
    'Parch': 'pais_filhos',
    'Ticket': 'bilhete',
    'Fare': 'tarifa',
    'Cabin': 'cabine',
    'Embarked': 'embarque'
}

caminho_arquivo = '/content/titanic.csv'

try:
    df = pd.read_csv(caminho_arquivo)
    print("✅ DataFrame 'df' carregado com sucesso.")

    df = df.rename(columns=mapa_colunas)

    print("✅ Colunas renomeadas para Português.")
    print("\nNovas Colunas:", df.columns.tolist())
    print("\nPrimeiras linhas para verificação:")
    print(df.head(2))

except FileNotFoundError:
    print("⚠️ ERRO: Arquivo não encontrado. Certifique-se de que a Célula 1 foi executada com sucesso.")
```

```python
def visualizar_extremos(df):
    print("\n--- VISUALIZAÇÃO DE EXTREMOS ---")
    print(df.head(4)) 

def mostrar_dimensoes(df):
    linhas, colunas = df.shape
    print(f"\n--- DIMENSÕES ---")
    print(f"O dataset possui: {linhas} Linhas e {colunas} Colunas.")

def mostrar_info_e_tipos(df):
    print("\n--- INFORMAÇÕES E TIPOS (df.info()) ---")
    df.info()

if not df.empty:
    visualizar_extremos(df)
    mostrar_dimensoes(df)
    mostrar_info_e_tipos(df)
    print("\nAnálise Inicial Concluída!")
```
