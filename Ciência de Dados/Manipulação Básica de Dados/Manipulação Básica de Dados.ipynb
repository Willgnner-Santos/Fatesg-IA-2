```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('imdb_top_1000.csv')
df.head()
```

```python
pd.concat([df.head(8), df.tail(8)])
```

```python
print(f"O dataset tem {df.shape[0]} linhas e {df.shape[1]} colunas")
```

```python
print(df.dtypes.to_frame(name='Tipo de Dado'))
```

```python
df["Released_Year"] = pd.to_numeric(df["Released_Year"], errors="coerce")

df = df.dropna(subset=["Released_Year"])

df["Released_Year"] = df["Released_Year"].astype(int)

step = 10  # intervalo de 10 anos

bins = range(
    df["Released_Year"].min(),
    df["Released_Year"].max() + step,
    step
)

df["Year_Bin"] = pd.cut(df["Released_Year"], bins=bins, right=False)

plt.figure(figsize=(12,5))
df["Year_Bin"].value_counts().sort_index().plot(kind="bar")
plt.title("Distribuio dos anos de lanamento por intervalos")
plt.xlabel("Intervalos de anos")
plt.ylabel("Frequncia")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

df["Runtime"] = df["Runtime"].astype(str).str.replace(" min", "", regex=False)
df["Runtime"] = pd.to_numeric(df["Runtime"], errors="coerce")


df = df.dropna(subset=["Runtime"])
df["Runtime"] = df["Runtime"].astype(int)

runtime_step = 20  # intervalo de 20 minutos

runtime_min = int(df["Runtime"].min())
runtime_max = int(df["Runtime"].max())

bins_runtime = range(runtime_min, runtime_max + runtime_step, runtime_step)

df["Runtime_Bin"] = pd.cut(df["Runtime"], bins=bins_runtime, right=False)

plt.figure(figsize=(12,5))
df["Runtime_Bin"].value_counts().sort_index().plot(kind="bar")
plt.title("Distribuio da durao dos filmes por intervalos")
plt.xlabel("Intervalos de durao (min)")
plt.ylabel("Frequncia")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist(df["IMDB_Rating"].dropna(), bins=15)
plt.xlabel("IMDB Rating")
plt.ylabel("Frequncia")
plt.title("Distribuio das notas IMDB")
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.hist(df["Released_Year"].dropna(), bins=20)
plt.xlabel("Ano de lanamento")
plt.ylabel("Frequncia")
plt.title("Distribuio dos anos de lanamento")
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.hist(df["Runtime"].dropna(), bins=20)
plt.xlabel("Durao (min)")
plt.ylabel("Frequncia")
plt.title("Distribuio das duraes dos filmes")
plt.show()
```

```python
df["Primary_Genre"] = df["Genre"].apply(lambda x: x.split(",")[0].strip())

plt.figure(figsize=(12,6))
df["Primary_Genre"].value_counts().head(15).plot(kind="bar")
plt.xlabel("Gnero")
plt.ylabel("Frequncia")
plt.title("Top 15 gneros mais frequentes")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df["Runtime"], df["IMDB_Rating"])
plt.xlabel("Durao (min)")
plt.ylabel("IMDB Rating")
plt.title("Durao vs Nota IMDB")
plt.show()
```
